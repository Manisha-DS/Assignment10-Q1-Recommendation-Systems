**Import Libraries**

```python
import pandas as pd
import numpy as np
```

**Import Dataset**

```python
df=pd.read_csv("book.csv",encoding='latin-1')
df
```

```python
# Renaming the column names
df1=df.rename({"Unnamed: 0":"Index","User.ID":"ID","Book.Title":"Title","Book.Rating":"Rating"},axis=1)
df1
```

```python
df2=df1.drop('Index',axis=True)
df2
```

# User Based Collaborative Filtering

```python
# Number of Unique ID's in Dataset
len(df2.ID.unique())
```

```python
# Number of Unique Titles in Dataset
len(df2.Title.unique())
```

**Pivot Table**

```python
# Creating Pivot Table
df3=df2.reset_index().pivot_table(index='ID',columns='Title',values='Rating').reset_index(drop=True)
df3
```

```python
df3.index=df2.ID.unique()
df3
```

```python
# Impute those NAN's with '0' values
df3.fillna(0, inplace=True)
df3
```

# Cosine Similarity

```python
# Calculating cosine similarity
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
sim = 1-pairwise_distances(df3.values,metric='cosine')
sim
```

```python
#Store the results in a dataframe
sim1= pd.DataFrame(sim)
#Set the index and column names to user ids 
sim1.index = df2.ID.unique()
sim1.columns = df2.ID.unique()
sim1
```

```python
np.fill_diagonal(sim, 0)
sim1.iloc[0:20,0:20]
```

```python
#Most Similar Users
sim1.idxmax(axis=1)[0:20]
```
